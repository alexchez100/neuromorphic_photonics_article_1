# Измерение спектра светодиода

```python
import numpy as np
import pandas as pd

file = r"C:\Users\alexc\Desktop\Data_boss\spe\905nm.txt"

data = pd.read_csv(file, sep = " ")
name_col = data.columns.tolist()
print (name_col, '\n')
print (data, '\n')
```

```python
x = data[name_col[1]]
y = data[name_col[2]]

print (x, '\n', y, '\n')
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15, 5)  # Размер картинок

fig, ax1 = plt.subplots()
ax1.plot(x, y, linewidth=2, label="905nm")

ax1.set_xlabel("Wavelength(nm)")
ax1.set_ylabel("Signal")
plt.title("Laser optical spectrum 905 nm")
ax1.legend()
ax1.axis([450, 1000, 0, 3500])
plt.show()
```

# Интенсивность на образце от расстояния LED до образца

```python
import numpy as np
import pandas as pd

file = r"C:\Users\alexc\Desktop\Data_boss\power.txt"

data = pd.read_csv(file, sep = "\t")
name_col = data.columns.tolist()
print (name_col, '\n')
print (data, '\n')
```

```python
x = data[name_col[0]]*10
y = data[name_col[1]]/1000/5/5*1000*1000

print (x, '\n', y, '\n')
```

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 5)  # Размер картинок

fig, ax1 = plt.subplots()
ax1.plot(x, y, linewidth=2, label="LED")

ax1.set_xlabel("Distance(mm)")
ax1.set_ylabel("Intensity(W/m^2)")
plt.title("Dependence of intensity on distance LED")
ax1.legend()
#ax1.axis([450, 1000, 0, 15000])
plt.show()
```
